```python
def sort(z,c=False,pad=0):
    up = int(z/2)-pad
    down = int(z/2)+pad

    zU, zL, g = slice(up,z), slice(0,down), slice(None)
    if type(c) != bool:
        cU, cL = slice(up,c), slice(0,down)
        return [(g,cU,zU),(g,cL,zU),(g,cU,zL),(g,cL,zL)] # Upper z priority, upper c 2nd priority
    return [(g,g,zU),(g,g,zL)]
```

```python
def get_slices(z, c=None, pad=0):
    up = int(z/2) - pad
    down = int(z/2) + pad

    zU, zL, g = slice(up, z), slice(0, down), slice(None)
    
    if c is not None:
        c_up = int(c/2) - pad
        c_down = int(c/2) + pad
        cU, cL = slice(c_up, c), slice(0, c_down)
        return [(g, cL, zL), (g, cU, zL), (g, cL, zU), (g, cU, zU)]  # Order matters for your priority
    
    return [(g, g, zL), (g, g, zU)]
```
